```python
data = torch.rand((4, 3, 224, 224)).cuda()
```

```python
start = torch.cuda.Event(enable_timing=True)
end = torch.cuda.Event(enable_timing=True)

times = []
for m, n, w, acc in models:
    mod = m(w).cuda()
    print(n)
    
    d = data
    if n == 'maxvit_t':
        d = torchvision.transforms.Resize((224, 224))(data)
    elif n == 'inception_v3':
        d = torchvision.transforms.Resize((299, 299))(data)
    elif n == 'vit_b_16':
        d = torchvision.transforms.Resize((384, 384))(data)
    elif n == 'vit_l_16':
        d = torchvision.transforms.Resize((512, 512))(data)
    
    start.record()

    for i in range(100):
        mod(d)

    end.record()
    torch.cuda.synchronize()
    time = start.elapsed_time(end)
    
    times.append((n, time, acc))
```

```python
df = pd.DataFrame(times, columns=['name', 'time', 'acc'])
df.sort_values('time')
```

```python
df.sort_values('time').loc[df['acc'] > 80]
# convnext_tiny is the best
```
